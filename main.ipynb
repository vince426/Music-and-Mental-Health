### Music & Mental Health Survey Results

Music therapy, or MT, is the use of music to improve an individual's stress, mood, and overall mental health. MT is also recognized as an evidence-based practice, using music as a catalyst for "happy" hormones such as oxytocin.

However, MT employs a wide range of different genres, varying from one organization to the next.

The MxMH dataset aims to identify what, if any, correlations exist between an individual's music taste and their self-reported mental health. Ideally, these findings could contribute to a more informed application of MT or simply provide interesting sights about the mind.

Thanks to Catherine Rasgaitis for this dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('musicmentalhealth.csv')
```

```python
data.head()
```

### Visualizations

### Pie Chart

Distribution of Primary Streaming Service

```python
colors = ['#90B494', '#FFA836', '#343434']
fig = plt.figure(facecolor='#172e08')
data['Primary streaming service'].value_counts().plot(kind='pie', figsize=(6,6), colors=colors, autopct='%1.1f%%', textprops={'color': 'white'})
```

### Bar Chart

Distribution of favorite genre

```python
fig = plt.figure(facecolor='#172e08')
bar_plot = data['Fav genre'].value_counts().plot(kind='bar', figsize=(14,6), color='#90B494').set_facecolor('#172e08')

plt.xlabel('Favorite Genre', color='white')
plt.ylabel('Count', color='white')
plt.title('Favorite Genre Distribution', color='white')

ax = plt.gca()
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')

ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
```

### Scatterplots

Finding correlation between number of hours listenes vs mental health issues

```python
fig, ax = plt.subplots(figsize=(14,6))
data.plot(kind='scatter', y='Hours per day', x='Anxiety', ax=ax, color='white')

ax.set_facecolor('#172e08')
fig.set_facecolor('#172e08')

ax = plt.gca()
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')

ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

plt.xlabel('Anxiety', color='white')
plt.ylabel('Hours per Day', color='white')
```
